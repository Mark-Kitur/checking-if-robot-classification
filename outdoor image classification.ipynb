# Preprocess data

```python
import tensorflow as tf
```

```python
IMG_SIZE = 224
def preprocess_image(fp):
    image = tf.io.read_file(fp)
    image =tf.image.decode_jpeg(image,channels=3)
    image =tf.image.convert_image_dtype(image,tf.float32)
    image = tf.image.resize(image,size=[IMG_SIZE,IMG_SIZE])
    return image
```

```python
def get_labels (fp,label):
    image = preprocess_image(fp)
    
    return image, label
```

# Batching

```python
BATCH_SIZE  =32
def batching (X,Y=None,BATCH_SIZE=BATCH_SIZE,validation_data=False):
    if validation_data:
        print('Creating validation data batches set ....')
        
        data= tf.data.Dataset.from_tensor_slices((tf.constant(X),
                                                tf.constant(Y)))
        data_batch=data.map(get_labels).batch(BATCH_SIZE)
        return data_batch
        
    else:
        print("Creating training data batches set....")
        data= tf.data.Dataset.from_tensor_slices((tf.constant(X),
                                                tf.constant(Y)))
        data= data.shuffle(buffer_size=len(X))
        data= data.map(get_labels)
        data_batch= data.batch(BATCH_SIZE)
        return data_batch
```

```python
train_data= batching(x_train, y_train)
```

```python
val_data= batching(x_test, y_test, validation_data=True)
```

```python
val_data.element_spec
```

# Build the model

```python
import tensorflow_hub as hub
```

```python
INPUT_SHAPE = [None,224,224,3]
OUTPUT_SHAPE =8
MODEL_URL ="https://www.kaggle.com/models/google/efficientnet-v2/frameworks/TensorFlow2/variations/imagenet21k-ft1k-b0-classification/versions/1"
```

```python
def build_model():
    model = tf.keras.Sequential([hub.KerasLayer(MODEL_URL),
                                tf.keras.layers.Dense(units=OUTPUT_SHAPE,
                                                    activation ='sigmoid')])
    model.compile(
        loss = tf.keras.losses.BinaryCrossentropy(),
        optimizer=tf.keras.optimizers.Adam(),
        metrics=['accuracy']
    )
    model.build(INPUT_SHAPE)
    return  model
```

```python
model=build_model()
```

```python
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='accuracy',
                                                patience=1)
```

```python
model.fit(x=train_data,
        validation_data=val_data,
        validation_freq=1,
        epochs=20)
```

```python
model.save('mymodel.h5')
```
